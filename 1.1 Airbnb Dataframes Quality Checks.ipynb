## 03.3 Changing Variable's Data Type.

## 03.3.1 'bnblist' dataframe.

#### Changing teh following variables' data types:

#### 'bedrooms' from float64 to Int64
#### 'price from object to Int64

#### bnblist['bedrooms'] = bnblist['bedrooms'].astype('Int64');
#### bnblist['price'] = bnblist['price'].astype('Int64')

#### This code was used to change both variables' data types to Int64.
#### In the 'bedrooms' variable case, it was possible to keep both Int64 values and  NaN values--see cell below.
#### The 'price' variable must be derived to contain only Int64 elements and take off the string elemts such as "$" sign.

```python
bnblist['bedrooms'] = bnblist['bedrooms'].astype('Int64')
```

```python
bnblist['price'] = bnblist['price'].astype('Int64')
```

#### For the 'price' variable, the '$' sign must be taken off of each one of its values.

```python
# Removing with strip function '$' symbol from 'price' variable:

[x.strip('$') for x in bnblist.price]
```

```python
# Applying changes to 'price' variable:

bnblist.price = [x.strip('$') for x in bnblist.price]
```

```python
# Removing ',' (comma) from 'price' variable:

bnblist.price.apply(lambda x: x.replace(',', ''))
```

```python
# Applying changes to 'price' variable: 

bnblist.price = bnblist.price.apply(lambda x: x.replace(',', ''))
```

```python
bnblist['price'].astype('float64')
```

```python
# Converting 'price' variable from object to float to int:

bnblist['price'] = bnblist['price'].astype('float').astype('int')
```
